```python
import numpy as np

def simulation(period, war, coun, seed):
    cost = 0
    country_list = {'before': {'USA': [8, 1.5], 'Saudi': [6, 0.5], 'Russia': [5, 0.5]},
                    'after': {'USA': [8, 2], 'Saudi': [6, 1.5], 'Russia': [7, 4]}}
    cost_list = [0]
    # 1 period = 5일, 6 period = 한 달 로 진행됩니다.
    oil_price_list = {'before': {'USA': [62.11, 66.10, 66.96, 68.73, 73.34],
                                 'Saudi': [65.28, 65.68, 69.02, 73.06, 74.97],
                                 'Russia': [65.93, 65.14, 67.52, 71.12, 76.26]},
                      'after': {'USA': [97.07, 104.40, 107.70, 117.75, 119.27],
                                'Saudi': [105.96, 109.02, 112.05, 119.90, 114.38],
                                'Russia': [99.12, 111.70, 110.84, 96.29, 89.88]}}
    oil_price_print = []
    demand = [7] * period
    inventory = [100]
    order = []
    order_time = [0] * period
    arrival_time = ['-'] * period
    lead_time = ['-'] * period

    if war == 'before':
        inventory_cost = 8
    else:
        inventory_cost = 10

    lead_mean, lead_std = country_list[war][coun]
    np.random.seed(int(seed))
    for i in range(period):
        oil_price = oil_price_list[war][coun][int(i / 6)]
        print("<Period %i>" % i)
        oil_price_print.append(oil_price)
        for sample_period, order_amount in order:
            if sample_period <= i:
                print("주문이 도착했습니다!!")
                inventory[i] += order_amount
                order.pop(0)
        print("현재 석유 가격: {}".format(oil_price))
        print("현재 수요: %i" % demand[i])
        print("현재 보관량: %i" % inventory[i])
        print("현재 비용: {}".format(cost))

        while True:
            print("행동을 선택하세요!")
            print("[O] 주문")
            print("[N] 다음 Period로")
            user = input()
            while (user != "O" and user != "N"):
                user = input()
            if user == "O":
                print("얼마나 주문하실 건가요?")
                while True:
                    user_order = input()
                    try:
                        user_order = int(user_order)
                        print("%i를 주문했습니다!" % user_order)
                        cost += user_order * oil_price
                        order_time[i] = 1
                        arrival = i + np.random.normal(lead_mean, lead_std, 1)
                        lead_time[i] = int(arrival[0]) + 1 - i
                        arrival_time[i] = int(arrival[0]) + 1
                        order.append((arrival_time[i], user_order))
                        break
                    except:
                        continue

            print("================================")
            break
        if inventory[i] > 0:
            cost += (inventory[i]) * inventory_cost
        else:
            cost -= (inventory[i]) * (3 * oil_price)

        inventory.append(inventory[i] - demand[i])
        cost_list.append(cost)
    return cost_list[-1]


def competition():
    print("컴페티션 예선에 오신 걸 환영합니다.")
    print("*** Case 1. 전쟁 전 재고 관리 ***")
    print("나라를 선택해주세요")
    print("USA / Saudi / Russia 중에 선택해주세요")
    coun = input()
    while (coun != "USA" and coun != "Saudi" and coun != "Russia"):
        coun = input()

    while True:
        print("시드 번호를 입력하세요.")
        seed = input()
        while (not seed.isdigit()):
            seed = input()
        print("시드 번호 %i가 확실하면 Y를 입력하세요." % int(seed))
        answer = input()
        if answer == "Y":
            break
    print("시뮬레이션을 시작합니다.")
    print("===========================")
    result = simulation(25, "before", coun, seed)
    print("최종 소모 비용: %f" % result)
    print("==================================")
    print("다음 지시가 있기 전까지 아무 키도 입력하지 마세요.")
    input()


    print("*** Case 2. 전쟁 후 재고 관리 ***")
    print("나라를 선택해주세요")
    print("USA / Saudi / Russia 중에 선택해주세요")
    coun = input()
    while (coun != "USA" and coun != "Saudi" and coun != "Russia"):
        coun = input()

    while True:
        print("시드 번호를 입력하세요.")
        seed = input()
        while (not seed.isdigit()):
            seed = input()
        print("시드 번호 %i가 확실하면 Y를 입력하세요." % int(seed))
        answer = input()
        if answer == "Y":
            break
    print("시뮬레이션을 시작합니다.")
    print("===========================")
    result = simulation(25, "after", coun, seed)
    print("최종 소모 비용: %f" % result)
    print("==================================")


competition()
```
